# Data Assembly

Now we will combine the necessary parts of the two datasets into one.

## Filtering Dataset #1 for the 'Strategy' Genre

For now, considering the size of the primary dataset and the fact that our immediate next step is taking a subset of it based on a specific genre (which will reduce the dataframe size for subsequent analyses), we'll only bring in the genre columns from dataset 2 to merge with dataset 1 first. If, instead, we first bring in all the important columns from dataset 2 and *then* filter, that will be a more computationally expensive process to get to the same resulting dataframe.

Note that, this doesn't necessarily mean that any one game is categorized under one, and only one, genre (e.g., a strategy game can have action gameplay elements). For the purpose of this project, we do not think it is necessary to check if any one game is assigned one and only one genre as that will disregard, say, some strategy games' reviews if they had action elements being reviewed (when I'm strictly filtering in strategy category and filtering out action category), but we want to include action element reviews as well if they were part of strategy games.

### Creating 'Genre' Dataframe from Dataset #2 (to Help Filter Dataset #1 Later)

First, we create a `Genre` dataframe from dataset #2 with a view to helping filter dataset #1 later on.

```python
# Creating a dataframe from dataset 2 with only steam IDs and strategy genre confirmation columns
df2_genre = df2[['ResponseID', 'ResponseName', 'GenreIsStrategy']]

## Genre dataframe has 'ResponseID' while dataset #1 has 'steam_id' to denote steam IDs
## Therefore changing 'ResponseID'-->'steam_id'
df2_genre = df2_genre.rename(columns={'ResponseID': 'steam_id', 'ResponseName': 'title'})

print("Dataframe to identify steam IDs and confirmation of whether the game is a strategy game or not:")
df2_genre
```

### Joining 'Genre' dataframe with Dataset #1

We will join the `Genre` dataframe with dataset #1 in order to be able to filter for it.

```python
# Left-joining dataset #1 ID with the Genre dataframe on steam ID
df1_genre = df1.merge(df2_genre,
                      on='steam_id',
                      how='left')
```

```python
df1_genre
```

### Filtering Augmented Dataset #1 to Include Only Strategy Games Reviews

Now, we will filter dataset #1 based on the `GenreIsStrategy` column to include only strategy games.

```python
df1_strategy = df1_genre.loc[df1_genre['GenreIsStrategy'] == True]

print("Primary dataset filtered to include only strategy games:")
df1_strategy
```

We will confirm that this dataframe now includes only strategy games by checking what value(s) the `GenreIsStrategy` column is taking now:

```python
# Printing all unique values in the 'GenreIsStrategy' column
print(f"List of all values the 'GenreIsStrategy' column contains: {df1_strategy['GenreIsStrategy'].unique()}")
```

Now that we are sure, we can remove the `GenreIsStrategy` column as its purpose is served (and move the `title` column next to `steam_id` for better presentation of the dataframe):

```python
# Dropping the 'GenreIsStrategy' column
df1_strategy = df1_strategy.drop(columns=['GenreIsStrategy'])

# Reorganizing the column orders
df1_strategy = df1_strategy[['steam_id', 'title', 'review_text', 'recommended', 'num_helpful']]

# Displaying updated df1_strategy dataframe
df1_strategy
```
